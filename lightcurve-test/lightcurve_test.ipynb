```python
import lightkurve as lk
print("Lightkurve version:", lk.__version__)
```

```python
search_result = lk.search_lightcurve("Kepler-10", mission="Kepler")
print(search_result)
```

```python
lc = search_result[0].download()
print(type(lc))
print(lc)
```

```python
%matplotlib inline
lc.plot(title="Kepler-10 Raw Light Curve")
```
